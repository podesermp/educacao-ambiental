```python
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd
import seaborn as sns

import json
import string
```

```python
def load_data(file_name:string) -> pd.DataFrame:
    with open(file_name) as file:
        data = json.load(file)
    vazao = []
    for i in data['sensor_data']:
        vazao.append(i['vazaoL/m'])
    data={
        "time": range(1,len(vazao)+1),
        "vazaoL/m":vazao
    }
    df = pd.DataFrame(data)
    return df
```

```python
torneira_1 = load_data(file_name='torneira-1.json')
torneira_2 = load_data(file_name='torneira-2.json')
```

```python
torneiras = pd.concat([torneira_1['time'],torneira_1['vazaoL/m'], torneira_2['vazaoL/m']], axis=1)
torneiras.columns = ['time', 'torneira1', 'torneira2']
torneiras
```
